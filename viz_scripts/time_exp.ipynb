```python
import numpy as np 
import pylab 
```

```python
l2g = np.loadtxt("../data/03_24_time_l2g.txt").mean(axis=1)
mds = np.loadtxt("../data/03_24_time_mds.txt").mean(axis=1)
umap = np.loadtxt("../data/03_24_time_umap.txt").mean(axis=1)
tsne = np.loadtxt("../data/03_24_time_tsnet.txt").mean(axis=1)
tsne2 = np.loadtxt("../data/03_28_time_tsnet.txt").mean(axis=1)
x = list(range(200,4001,200))
```

```python
pylab.plot(x,l2g,'o-',label="L2G")
pylab.plot(x,mds, 'o-',label="MDS")
pylab.plot(x,umap,'o-',label="UMAP")
pylab.plot(x[:9],tsne[:9], 'o-',label="tsNET")
pylab.suptitle("")
pylab.ylabel("time in seconds")
pylab.xlabel("|V|")
pylab.yscale("log")
pylab.legend()
pylab.suptitle("Scalability of algorithms")
pylab.savefig("figures/new_time_exp.pdf")
```

```python
import numpy as np 
import pylab 
```

```python
old = np.loadtxt("../data/03_13_timeexp.txt")
new = np.loadtxt("../data/03_13_new_timeexp.txt")
x = list(range(200,4001,200))
```

```python
pylab.plot(x,old,'o-',label="JIT Python Implementation")
pylab.plot(x,new, 'o-',label="(New) C Implementation")
pylab.suptitle("Comparision of optimization implementations")
pylab.ylabel("time in seconds")
pylab.xlabel("|V|")
pylab.legend()
pylab.savefig("figures/old_new_opt_compare.png")
```
